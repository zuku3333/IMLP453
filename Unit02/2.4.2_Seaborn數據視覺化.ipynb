# #單變量分析繪圖（直方圖、條形圖）
> Dataset: NBA_season1718_salary.csv<br>
ref:https://www.kaggle.com/koki25ando/salary

```python
data = pd.read_csv('data/NBA_season1718_salary.csv')
data
```

```python
#對讀進來的數據按薪資降序排序並取前10條數據
salary_top10 = (data.sort_values('season17_18', ascending=False)).head(10)
salary_top10
```

```python
#取出球員列數據，作為條形圖橫坐標
player = salary_top10['Player']


#取出球員薪資列，作為條形圖縱坐標
season_salary = salary_top10['season17_18']
```

```python
plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
plt.ylim(season_salary.min() * 0.5, season_salary.max() * 1.01)

# 繪製條形圖（ci=68表示信賴區間68%，約為1個標準差）
sns.barplot(x=player, y=season_salary, data=salary_top10, ci=68)
#ci參數表示允許的誤差範圍（控制誤差棒的百分比，在0-100之間）
plt.show()
```

## Result

從條形圖中不難看出，勇士隊的當家球星curry資最高，將近達到3500萬美元，位居第二的是勒布朗詹姆斯，也是在3300萬美元以上。上面的條形圖是按球員作為橫坐標，下面我們按球隊匯總，看看哪個球隊的薪資總額最高。

# 練習 - 彙總球隊薪資繪製成直方圖
按球隊匯總，看看哪個球隊的薪資總額最高。

```python
# 第一段：按隊伍分組計算薪資總和
salary_by_team = data.groupby(by='Tm').sum()
salary_by_team
```

```python
# 先將資料集用groupby根據球隊分別匯總
# 按隊伍分組並重設索引
salary_by_team = data.groupby(by='Tm').sum().reset_index()

# 按薪資降序排列
salary_by_team = salary_by_team.sort_values(by='season17_18', ascending=False)

# 取前10隊
tm_top10 = salary_by_team[1:10]

# 提取隊名和薪資
tm = tm_top10['Tm']
salary_sum = tm_top10['season17_18']


# 設定調色盤和樣式
colors = sns.color_palette('husl', 10)
sns.set(style='whitegrid')

# 繪製長條圖
sns.barplot(x=tm, y=salary_sum, data=tm_top10, palette=colors)
plt.show()
```

```python
# 匯入統計模組
import scipy.stats as stats

# 產生伽瑪分布資料
x = np.random.gamma(6, size=200)

# 繪製分布圖
sns.distplot(x, kde=False, fit=stats.gamma)

'''
kde=False：不顯示核密度估計曲線
fit=stats.gamma：擬合伽瑪分布曲線到資料上
結果：顯示直方圖 + 理論分布曲線
'''

# 多變量常態分布
mean, cov = [0, 1], [[1, .5], [.5, 1]]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=['x', 'y'])
'''
mean = [0, 1]：兩個變數的平均值
cov = [[1, .5], [.5, 1]]：共變異數矩陣
0.5 > 0：x 和 y 正相關
'''

# 聯合分布圖
sns.jointplot(x='x', y='y', data=df)

plt.show()
```

# #繪製盒圖、小提琴圖
seaborn.boxplot()
這個函數主要是繪製出一個箱型圖來反映離群點數據。

```python
sns.boxplot(x='tip',data=tips)
plt.show()
```

```python
sns.boxplot(y='tip',data=tips)
plt.show()
```

```python
sns.boxplot(y='tip',x='day',data=tips,hue='day')
plt.show()
```

seaborn庫中的很多繪圖方法就是提前給你畫好模板，我們只需要把數據傳進去就得到相應的圖形，
基於此，我門是不是可以在同一個模板上同時繪製分簇圖和箱型圖呢？

```python
sns.boxplot(y='total_bill',x='day',data=tips, hue='day')
sns.swarmplot(y='total_bill',x='day',data=tips, color='.25')
plt.show()
```

```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='day')
plt.show()
```

小提琴圖即反映了數據的離群情況，同時也反映了數據的分布密度。<br>
此函數跟箱型圖一樣，也有hue參數，可以查看第二個分類屬性下的數據分布，如下：


```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex')
plt.show()
```

可以看到，上面的小提琴圖是4對8個，即當設置hue參數為sex時，每個小提琴的數據集是一類數據，那有沒有辦法讓男性對應數據和女性對應數據都作為一個數據集繪製出一個小提琴圖，只是在一個圖中區分男性數據和女性數據呢？答案是肯定的，violinplot()函數提供了一個split參數，可以做到以上需求，如下：


```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex',split=True)
plt.show()
```

除此以外，我們再來看一下inner參數，上面的小提琴圖中內部都是箱型圖，通過inner參數，我們可以設置其他類型，inner可選參數有：box、quartile(四分位)、point、stick。我們依次看下：


```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex',split=True,inner='box')
plt.show()
```

```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex',split=True,inner='quartile')
plt.show()
```

```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex',split=True,inner='point')
plt.show()
```

```python
sns.violinplot(y='total_bill',x='day',data=tips, hue='sex',split=True,inner='stick')
plt.show()
```

# #結構化展示多維數據
***Visualizing the multidimensional relationships among the samples is as easy as calling sns.pairplot:***

```python
sns.pairplot(tips, hue='sex', height=2.5)
plt.show()
```

使用FacetGrid時，我們會通過一個pandas DataFrame以及控制圖形網格的行、列和顏色的變量名稱來初始化一個對象。這些維度變量（控制行、列和顏色的變量）應該是分類變量或者離散變量，然後這些變量的不同水平組合起來就構成了整個圖形的每一個子圖（facet，在這裡可以理解為我們維度拆解的最小粒度）


這一函數的目標是一步到位地提供一幅完整的成品圖，它在完成繪圖後還會對每個坐標軸添加注釋。

```python
g = sns.FacetGrid(tips,col='time')
plt.show()
```

```python
g.map(plt.hist,'tip')
plt.show()
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')          # 按性別分欄，按吸菸分色
g.map(plt.scatter, 'total_bill', 'tip', alpha=.7)         # 繪製散點圖
g.add_legend()                                            # 添加圖例
plt.show()
```

我們可以指定某個seaborn調色板，也可以通過字典將hue變量中的每個分類與其對應的matplotlib顏色傳遞給函數<br>
（這樣就可以隨心所以使用大量的matplotlib支持的色彩）


```python
pal = dict(Lunch='seagreen', Dinner='gray')               # 自訂顏色字典
g = sns.FacetGrid(tips, hue='time', palette=pal, height=5) # 按用餐時間分色
g.map(plt.scatter, 'total_bill', 'tip', s=50, alpha=.7, linewidth=.5, edgecolor='white')
g.add_legend()
plt.show()

# - 大小為 50 的圓點
# - 70% 不透明（有點半透明）
# - 點的白色細邊框（寬度 0.5）
# - 點邊框顏色由 hue 參數決定
```

```python
tips = sns.load_dataset('iris')
g = sns.PairGrid(tips)            # 建立配對網格
g.map(plt.scatter)                # 所有變數兩兩配對散點圖
plt.show()
```

map_diag()函數是繪製對角線上的單變量子圖，mag_offdiag是繪製對角線以外的兩個變量間的子圖。

```python
g = sns.PairGrid(tips)
g = g.map_diag(plt.hist, edgecolor='w')              #對角線位置（變數對自己）：使用直方圖
g = g.map_offdiag(plt.scatter, edgecolor='w', s=40)  #非對角線位置（不同變數間）：使用散點圖
plt.show()
```

# [練習]試著讀取自己的資料，繪製出多維數據圖表

```python
tips = sns.load_dataset('titanic')
numeric_cols = ['age', 'fare', 'parch', 'sibsp']
g = sns.PairGrid(tips[numeric_cols].dropna())  # 移除缺失值
g = g.map_diag(plt.hist, color='skyblue',edgecolor='w', alpha=0.7)              
g = g.map_upper(plt.scatter, color='red', s=20, edgecolor='w', alpha=0.7)
g = g.map_lower(plt.scatter, color='green', s=20, edgecolor='w', alpha=0.7) 
plt.show()
```
